```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
X = pd.DataFrame(np.loadtxt('X-DTLZ1.txt'))
Y = pd.DataFrame(np.loadtxt('Y-DTLZ1.txt'))
Y.columns = ['Var_' + str(i) for i in range(Y.shape[1])]
X.columns = ['Obj_' + str(i) for i in range(X.shape[1])]
X['Obj_0'] = -X['Obj_0']
```

```python
for j in range(Y.shape[1]):
    for i in range(X.shape[1]):
        obj = 'Obj_' + str(i)
        var = 'Var_'+ str(j)
        data = pd.concat([Y[var],X[obj]], axis=1)
        data.plot.scatter(x=obj, y=var, ylim=(0,1), xlim=(0, 1.5));
        plt.show()
```

```python
for i in range(Y.shape[1]):
    var1 = 'Var_0'
    var2 = 'Var_' + str(i)
    data = pd.concat([Y[var1],Y[var2]], axis=1)
    data.plot.scatter(x=var1, y=var2, ylim=(0, 1.5));
```

```python
data = pd.concat([X,Y], axis=1)
corrmat = data.corr()
f, ax = plt.subplots(figsize=(12, 9))
sns.heatmap(corrmat, vmax=.8, square=True)
```

```python
data = X[X.Obj_0]
data.Obj_0
sns.distplot(data['Obj_0']);
```

```python
sns.set()
data = pd.concat([X['Obj_0'],Y], axis=1)
sns.pairplot(data, height = 2.5)
plt.show();
```
